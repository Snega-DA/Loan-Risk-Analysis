### Good Loan Metrics

```python
good_loan=df[df["loan_status"].isin(["Fully Paid","Current"])]

good_loan_applications = good_loan["id"].count()
good_loan_funded_amount = good_loan["loan_amount"].sum()/1000000
good_loan_received = good_loan["total_payment"].sum()/1000000

good_loan_percentage = (good_loan_applications/ total_loan_application) * 100

print("good_loan_applications:",good_loan_applications)
print("good_loan_funded_amount:{:.2f}M".format(good_loan_funded_amount))
print("good_loan_received:{:.2f}M".format(good_loan_received))
print("percentage_of_good_loan_applications:{:.2f}%".format(good_loan_percentage))
```

### Bad Loan Metrics

```python
bad_loan=df[df["loan_status"].isin(["Charged Off"])]

bad_loan_applications = bad_loan["id"].count()
bad_loan_funded_amount = bad_loan["loan_amount"].sum()/1000000
bad_loan_received = bad_loan["total_payment"].sum()/1000000

bad_loan_percentage = (bad_loan_applications/ total_loan_application) * 100

print("bad_loan_applications:",bad_loan_applications)
print("bad_loan_funded_amount:{:.2f}M".format(bad_loan_funded_amount))
print("bad_loan_received:{:.2f}M".format(bad_loan_received))
print("percentage_of_bad_loan_applications:{:.2f}%".format(bad_loan_percentage))
```

### Monthly Trend of Loan Applications

```python
monthly_applications = (
    df.sort_values('issue_date').assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y')).groupby('month_name', sort=False)['id']
      .count().reset_index(name='loan_applications_count'))

plt.figure(figsize=(10, 5))
x = monthly_applications['month_name']
y = monthly_applications['loan_applications_count']
plt.fill_between(x, y, color='navy', alpha=0.5)
plt.plot(x, y, color='lightblue', linewidth=2)
for i, v in enumerate(y):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=9)
plt.title('Total Loan Applications by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Number of Applications')
plt.xticks(range(len(x)), x, rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### Monthly Trends of Total Funded Amount

```python
monthly_funded = (df.sort_values("issue_date").assign(month_name=lambda x:x["issue_date"].dt.strftime("%b %y")).
                   groupby("month_name",sort=False)["loan_amount"].sum().div(1_00_000).
                   reset_index(name="loan_amount_millions"))

plt.figure(figsize=(10,5))
plt.fill_between(monthly_funded["month_name"], monthly_funded["loan_amount_millions"],color="lightgreen",alpha=0.5)
plt.plot(monthly_funded['month_name'],monthly_funded['loan_amount_millions'],color='green',linewidth=2)

for i, row in monthly_funded.iterrows():
    plt.text(i,row['loan_amount_millions'] + 0.1,f"{row['loan_amount_millions']:.2f}",ha='center',va='bottom',fontsize=9,rotation=0,color='black')

plt.title('Total Funded Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Funded Amount (Millions)')
plt.xticks(ticks=range(len(monthly_funded)),labels=monthly_funded['month_name'],rotation=45)
plt.tight_layout()
plt.show() 
```

### Monthly Trends of Total received Amount

```python
monthly_received= (df.sort_values("issue_date").assign(month_name=lambda x:x["issue_date"].dt.strftime("%b %y")).
                   groupby("month_name",sort=False)["total_payment"].sum().div(1000000).
                   reset_index(name="received_amount_millions"))

plt.figure(figsize=(10,5))
plt.fill_between(monthly_received["month_name"], monthly_received["received_amount_millions"],color="plum",alpha=0.5)
plt.plot(monthly_received['month_name'],monthly_received['received_amount_millions'],color='purple',linewidth=2)

for i, row in monthly_received.iterrows():
    plt.text(i,row['received_amount_millions'] + 0.1,f"{row['received_amount_millions']:.2f}",ha='center',va='bottom',fontsize=9,rotation=0,
             color='black')

plt.title('Total received Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Received Amount (Millions)')
plt.xticks(ticks=range(len(monthly_received)),labels=monthly_funded['month_name'],rotation=45)
plt.tight_layout()
plt.show()  
```

### Loan Term by Funded Amount

```python
term_funding_millions = df.groupby('term')['loan_amount'].sum() / 1_000_000
total = term_funding_millions.sum()
plt.figure(figsize=(5, 5))
plt.pie(
    term_funding_millions,
    labels=term_funding_millions.index,
    autopct=lambda p: f"{p:.1f}%\n{(p*total/100):.1f}M",
    startangle=90,
    colors=['maroon','mistyrose'])
plt.title('Total Funded Amount by Term (in $ Millions)')
plt.show()
```

### Loan Purpose by Funded Amount

```python
p = df.groupby('purpose')['loan_amount'].sum().sort_values() / 1_000_000

plt.figure(figsize=(10, 5))
bars = plt.bar(p.index, p.values, color='seagreen')
plt.ylim(0, p.max() * 1.15) 
for i, v in enumerate(p.values):
    plt.text(i, v + p.max()*0.02, f"{v:.2f}M", ha='center', fontsize=9)
plt.title('Total Funded Amount by Loan Purpose (₹ Millions)', fontsize=14)
plt.xlabel('Loan Purpose')
plt.ylabel('Funded Amount (₹ Millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Total Received Amount By Employment Length

```python
emp_repayment = df.groupby('emp_length')['total_payment'].sum().sort_values() / 1000

plt.figure(figsize=(10, 6))
bars = plt.barh(emp_repayment.index, emp_repayment, color='pink')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 5, bar.get_y() + bar.get_height() / 2,
             f"{width:,.0f}K", va='center', fontsize=9)

plt.xlabel("Received Amount (₹ Thousands)")
plt.title("Total Received Amount by Employment Length")
plt.tight_layout()
plt.show()
```

### Funded Amount By State

```python
state_funding = df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=True)
state_funding_thousands = state_funding / 10000

plt.figure(figsize=(10, 8))
bars = plt.barh(state_funding_thousands.index,state_funding_thousands.values,color='lightcoral')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 10,bar.get_y() + bar.get_height() / 2,f'{width:.0f}K',va='center',fontsize=9)

plt.title('Total Funded Amount by State (in ₹ Thousands)')
plt.xlabel('Funded Amount (₹ \'000)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

### Charge - Off Rate by Grade

```python
df['is_charged_off'] = df['loan_status'].apply(lambda x: 1 if x == 'Charged Off' else 0)
grade_chargeoff = df.groupby('grade')['is_charged_off'].mean().sort_index()

plt.figure(figsize=(8,5))
grade_chargeoff.plot(kind='bar',color="teal")

plt.title("Charge-Off Rate by Grade")
plt.xlabel("Loan Grade")
plt.ylabel("Charge-Off Rate")
plt.xticks(rotation=0)
plt.show()
```

### Charge-Off Rate by Loan purpose

```python
df['is_charged_off'] = df['loan_status'].apply(lambda x: 1 if x == 'Charged Off' else 0)

purpose_default = (
    df.groupby('purpose')['is_charged_off'] .mean().sort_values())

plt.figure(figsize=(10,6))
purpose_default.plot(kind='barh',color="orchid")
plt.title("Charge-Off Rate by Loan Purpose")
plt.xlabel("Charge-Off Rate")
plt.ylabel("Purpose")
plt.show()
```

### Average DTI By Loan Status

```python
import seaborn as sns
avg_dti = df.groupby('loan_status')['dti'].mean()
sns.heatmap(avg_dti.to_frame(), annot=True, cmap='coolwarm')
plt.title("Average DTI by Loan Status")
plt.show()
```

### Home Ownership vs Risk

```python
home_risk = df.groupby('home_ownership')['loan_status'] \
    .apply(lambda x: (x == 'Charged Off').mean()*100)

avg_dti = df.groupby('home_ownership')['dti'].mean()
fig, ax1 = plt.subplots(figsize=(8,5))
bars = ax1.bar(home_risk.index, home_risk.values, 
               color='teal', label='Charge-Off Rate (%)')
ax1.set_ylabel("Charge-Off Rate (%)", color='teal')
ax2 = ax1.twinx()

line = ax2.plot(avg_dti.index, avg_dti.values, 
                color='gold', marker='o', label='Average DTI')
ax2.set_ylabel("Average DTI", color='gold')

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
plt.legend(handles1 + handles2, labels1 + labels2, loc='upper right')

plt.title("Home Ownership vs Risk & Financial Stress")
plt.xticks(rotation=45)

plt.show()
```
